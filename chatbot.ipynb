```python
from __future__ import annotations

# =============================================================================
# Travel Agency Chatbot
# =============================================================================
#
# Cél:
# - Egy európai utazási iroda chatbotjának megvalósítása.
# - A bot több témát kezel: tengerpart, hegy, kulturális városnézés,
#   extrém/kalandos utak, szállás, étkezés, biztosítás, megközelíthetőség,
#   időjárás, árak, foglalás és utazási időtartam.
#
# Megközelítés:
# 1) Szöveg normalizálása: egységesítjük a bemenetet, hogy a modell stabilabb legyen.
# 2) Intent osztályozó: TF-IDF + logisztikus regresszió.
# 3) Retrieval fallback: FAQ jellegű kérdés-válasz keresés TF-IDF hasonlósággal.
# 4) Hibabiztos hibrid router: szabályalapú gyors útvonal + ML + retrieval.
#
# Miért így?
# - A klasszikus ML gyors, könnyen értelmezhető és kevés erőforrást igényel.
# - A retrieval rész segít, ha a kérdés egy konkrét FAQ-hoz közel áll.
# - A hibrid logika jobb, mint egyetlen technika önmagában, mert robusztusabb.
#
# A kód úgy készült, hogy könnyen beilleszthető legyen Google Colab-ba is.
# =============================================================================

import re
import random
from dataclasses import dataclass
from typing import Dict, Iterable, List, Sequence, Tuple

import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split


# =============================================================================
# Reprodukálhatóság
# =============================================================================
# Ha később újra futtatjuk a notebookot/scriptet, ugyanazt az eredményt szeretnénk
# kapni a véletlenszerű műveleteknél (train/test split, stb.).
random.seed(42)
np.random.seed(42)


# =============================================================================
# 1) Adatkészlet: intentek és példamondatok
# =============================================================================
# Az intentek olyan felhasználói szándékok, amelyeket a chatbot felismer.
# A dataset kicsi, oktatási célú, de jól mutatja a teljes folyamatot.
RAW_DATA: List[Tuple[str, str]] = [
    # Köszönés / nyitás
    ("Szia, tudsz segíteni?", "greeting"),
    ("Helló, van valaki?", "greeting"),
    ("Jó napot, utazást keresek.", "greeting"),
    ("Üdv, milyen ajánlataitok vannak?", "greeting"),

    # Tengerparti utak
    ("Tengerparti nyaralást szeretnék.", "beach"),
    ("Van mediterrán tengerparti út?", "beach"),
    ("Görög vagy horvát tengerpart érdekel.", "beach"),
    ("Napfényes strandos utat keresek.", "beach"),

    # Hegyi utak
    ("Hegyi túrát szeretnék.", "mountain"),
    ("Van alpesi vagy dolomitokbeli út?", "mountain"),
    ("Szeretem a havas hegyeket.", "mountain"),
    ("Túrázós hegyi nyaralást keresek.", "mountain"),

    # Kulturális utak
    ("Kulturális városnézés érdekel.", "cultural"),
    ("Múzeumok és történelmi látnivalók kellenek.", "cultural"),
    ("Prága vagy Bécs jó lenne.", "cultural"),
    ("Szeretnék egy kulturális európai utat.", "cultural"),

    # Kaland / extrém sport
    ("Extrém sportos utat szeretnék.", "adventure"),
    ("Bungee jumping vagy rafting van?", "adventure"),
    ("Kalandos aktív nyaralást keresek.", "adventure"),
    ("Szeretek adrenalin programokat.", "adventure"),

    # Szállás
    ("Milyen szálláslehetőségek vannak?", "accommodation"),
    ("Hotel vagy apartman is van?", "accommodation"),
    ("All inclusive szállást keresek.", "accommodation"),
    ("Családbarát szállás érdekel.", "accommodation"),

    # Étkezés
    ("Az étkezés benne van az árban?", "food"),
    ("Félpanzió vagy all inclusive van?", "food"),
    ("Vegetáriánus étkezés megoldható?", "food"),
    ("Milyen étkezési opciók vannak?", "food"),

    # Biztosítás
    ("Utazási biztosítás van az ajánlatban?", "insurance"),
    ("Mire terjed ki a biztosítás?", "insurance"),
    ("Kell külön biztosítást kötnöm?", "insurance"),
    ("Balesetbiztosítás is elérhető?", "insurance"),

    # Megközelíthetőség
    ("Könnyen megközelíthető úticél kell.", "accessibility"),
    ("Van repülővel jól elérhető hely?", "accessibility"),
    ("Vonattal is elérhető az út?", "accessibility"),
    ("Gyerekekkel és idősekkel is jó legyen a megközelítés.", "accessibility"),

    # Időjárás / szezon
    ("Milyen az időjárás nyáron ott?", "weather"),
    ("Kora ősszel is jó az idő?", "weather"),
    ("Mikor a legjobb az időjárás tengerpartra?", "weather"),
    ("Hegyvidéken mennyire hideg van?", "weather"),

    # Úticélok
    ("Milyen európai úticélokat ajánlasz?", "destination"),
    ("Melyik országokat szervezitek?", "destination"),
    ("Van ajánlat Olaszországba vagy Spanyolországba?", "destination"),
    ("Érdekelnek a fő európai úticélok.", "destination"),

    # Ár / költség
    ("Mennyibe kerül ez az út?", "price"),
    ("Van kedvezmény vagy akció?", "price"),
    ("Melyik csomag a legolcsóbb?", "price"),
    ("Az ár tartalmazza a repülőt?", "price"),

    # Időtartam
    ("Hány napos az utazás?", "duration"),
    ("Rövid hétvégi utat keresek.", "duration"),
    ("Van egyhetes és kéthetes csomag?", "duration"),
    ("Mennyi ideig tart az út?", "duration"),

    # Foglalás
    ("Hogyan tudok foglalni?", "booking"),
    ("Mikor kell befizetni az előleget?", "booking"),
    ("Lehet online foglalni?", "booking"),
    ("Foglalást szeretnék intézni.", "booking"),

    # Közlekedés / transzfer
    ("Hogyan lehet oda eljutni?", "transport"),
    ("A repülőjegy benne van?", "transport"),
    ("Buszos utazás is van?", "transport"),
    ("Transzfert biztosítotok?", "transport"),

    # Család
    ("Családi nyaralást keresek.", "family"),
    ("Gyerekbarát programok vannak?", "family"),
    ("Kisgyerekkel is alkalmas az út?", "family"),
    ("Családos utazást ajánlanál?", "family"),
]


def build_dataframe(data: Sequence[Tuple[str, str]]) -> pd.DataFrame:
    # A DataFrame táblázatos formában kezelhetővé teszi az adatokat.
    # Ez különösen hasznos, ha később bővíteni vagy ellenőrizni akarjuk az adatokat.
    return pd.DataFrame(data, columns=["text", "intent"])


df = build_dataframe(RAW_DATA)


# =============================================================================
# 2) Szöveg-normalizálás
# =============================================================================
# A chatbot teljesítménye sokat javulhat, ha egységesítjük a bemenetet.
# Például:
# - kisbetűsítés
# - URL-ek eltávolítása
# - számok egységesítése
# - felesleges írásjelek eldobása
# - whitespace normalizálása
#
# Ez a lépés azért fontos, mert a TF-IDF modell egyszerű szóelőfordulásokból építkezik.
def basic_normalize_hu(text: str) -> str:
    t = text.lower().strip()
    t = re.sub(r"https?://\S+|www\.\S+", " ", t)
    t = re.sub(r"\d+", " <NUM> ", t)
    t = re.sub(r"[^a-záéíóöőúüű0-9\s<>]+", " ", t)
    t = re.sub(r"\s+", " ", t).strip()
    return t


def simple_tokenize(text: str) -> List[str]:
    # Egyszerű tokenizálás: a normalizált szöveget szóköz mentén feldaraboljuk.
    # Oktatási célra elég, bár valós NLP-projektekben ennél összetettebb tokenizer is kellhet.
    return basic_normalize_hu(text).split()


def make_ngrams(tokens: Sequence[str], n: int = 2) -> List[str]:
    # n-gram képzés: egymást követő tokenekből összeállítunk 1, 2, 3... hosszú egységeket.
    # Ez segíthet olyan kifejezések felismerésében, mint "mikor érkezik" vagy "jelszó elfelejtve".
    return [" ".join(tokens[i : i + n]) for i in range(len(tokens) - n + 1)]


# =============================================================================
# 3) Modell felépítése: TF-IDF + logisztikus regresszió
# =============================================================================
# A TF-IDF a gyakori, kevésbé informatív szavak súlyát csökkenti, a fontosabbakét emeli.
# A logisztikus regresszió egyszerű, gyors és jól működik kis, címkézett szövegkorpuszokon.
@dataclass(frozen=True)
class IntentPrediction:
    intent: str
    confidence: float
    probabilities: Dict[str, float]


class TravelIntentModel:
    def __init__(self) -> None:
        self.vectorizer = TfidfVectorizer(
            preprocessor=basic_normalize_hu,
            ngram_range=(1, 2),
            min_df=1,
        )
        self.classifier = LogisticRegression(max_iter=2000)
        self.is_fitted = False

    def fit(self, texts: Sequence[str], labels: Sequence[str]) -> None:
        X_vec = self.vectorizer.fit_transform(texts)
        self.classifier.fit(X_vec, labels)
        self.is_fitted = True

    def predict(self, text: str) -> str:
        if not self.is_fitted:
            raise RuntimeError("A modellt előbb tanítani kell a fit() metódussal.")
        X_vec = self.vectorizer.transform([text])
        return self.classifier.predict(X_vec)[0]

    def predict_with_confidence(self, text: str) -> IntentPrediction:
        if not self.is_fitted:
            raise RuntimeError("A modellt előbb tanítani kell a fit() metódussal.")
        X_vec = self.vectorizer.transform([text])
        probs = self.classifier.predict_proba(X_vec)[0]
        best_idx = int(np.argmax(probs))
        intent = self.classifier.classes_[best_idx]
        confidence = float(probs[best_idx])
        probability_map = dict(zip(self.classifier.classes_, probs))
        return IntentPrediction(intent=intent, confidence=confidence, probabilities=probability_map)

    def top_terms_by_intent(self, top_k: int = 6) -> Dict[str, List[Tuple[str, float]]]:
        # Ez a függvény segít értelmezni a modellt.
        # Megmutatja, hogy az egyes intentekhez mely szavak/bigramok tartoznak leginkább.
        feature_names = np.array(self.vectorizer.get_feature_names_out())
        result: Dict[str, List[Tuple[str, float]]] = {}

        for i, intent in enumerate(self.classifier.classes_):
            coefs = self.classifier.coef_[i]
            top_idx = np.argsort(coefs)[-top_k:][::-1]
            result[intent] = [(feature_names[j], float(coefs[j])) for j in top_idx]
        return result


# =============================================================================
# 4) FAQ / retrieval alapú fallback
# =============================================================================
# Ha a klasszifikátor bizonytalan, akkor a kérdést összevetjük egy kis FAQ adatbázissal.
# Ez különösen hasznos, ha a felhasználó egy előre ismert kérdést fogalmaz meg kicsit másképp.
FAQ: List[Tuple[str, str]] = [
    (
        "Melyik a legjobb tengerparti úticél?",
        "Görögország, Horvátország, Spanyolország és Portugália nagyon népszerű tengerparti célok.",
    ),
    (
        "Van hegyi túrázós út?",
        "Igen, az Alpok, Dolomitok és a Tátra környékén is szervezünk túracsomagokat.",
    ),
    (
        "Milyen kulturális városokat ajánlotok?",
        "Prága, Bécs, Róma, Budapest és Krakkó a leggyakoribb kulturális úticélok.",
    ),
    (
        "Kell utasbiztosítás?",
        "Ajánlott az utasbiztosítás, különösen aktív vagy külföldi utaknál.",
    ),
    (
        "Lehet félpanziót kérni?",
        "Igen, a legtöbb útnál választható reggeli, félpanzió vagy teljes ellátás.",
    ),
    (
        "Van gyerekbarát program?",
        "Igen, családi utaknál rövidebb programot és gyerekbarát szállást is ajánlunk.",
    ),
    (
        "Hogyan történik a foglalás?",
        "Online érdeklődés után visszaigazolást és előlegfizetési információt küldünk.",
    ),
    (
        "A repülőjegy benne van az árban?",
        "Ez csomagtól függ, de több ajánlatunk tartalmazza a repülőjegyet is.",
    ),
    (
        "Mikor a legjobb tengerpartra utazni?",
        "Tengerpartra általában május és szeptember között a legjobb az idő.",
    ),
    (
        "Mikor jó a hegyekbe menni?",
        "Túrázni nyáron, síelni télen érdemes a hegyvidéki célpontokra.",
    ),
    (
        "Milyen szállásopciók vannak az utakhoz?",
        "Hotel, apartman, panzió, családbarát szállás és all inclusive is elérhető.",
    ),
    (
        "Lehet vegetáriánus vagy speciális étkezést kérni?",
        "Igen, előre jelzett igény esetén vegetáriánus és egyéb speciális étkezés is megoldható.",
    ),
    (
        "Milyen közlekedéssel lehet menni?",
        "Repülős, buszos, vonatos és egyéni megközelítésű csomagokat is tudunk ajánlani.",
    ),
    (
        "Van ár-érték arányban jó ajánlat?",
        "Igen, több szezonális ajánlatunk is van, amelyek jó ár-érték arányt képviselnek.",
    ),
    (
        "Milyen európai úticélokat szerveztek nyáron?",
        "Nyáron népszerű a mediterrán tengerpart, az alpesi túra és a városnéző kulturális út is.",
    ),
]


class RetrievalEngine:
    def __init__(self, faq_pairs: Sequence[Tuple[str, str]]) -> None:
        self.questions = [q for q, _ in faq_pairs]
        self.answers = [a for _, a in faq_pairs]
        self.vectorizer = TfidfVectorizer(preprocessor=basic_normalize_hu, ngram_range=(1, 2))
        self.matrix = self.vectorizer.fit_transform(self.questions)

    def answer(self, text: str, min_similarity: float = 0.20) -> str:
        query_vec = self.vectorizer.transform([text])
        # A mátrixszorzat gyors hasonlóság-számítás TF-IDF vektorok között.
        similarities = (query_vec @ self.matrix.T).toarray()[0]
        best_idx = int(np.argmax(similarities))
        best_similarity = float(similarities[best_idx])

        if best_similarity < min_similarity:
            return "Nem találtam biztos találatot a tudásbázisban. Tudod pontosítani?"
        return f"(találat={best_similarity:.2f}) {self.answers[best_idx]}"


# =============================================================================
# 5) A bot válaszai intentenként
# =============================================================================
INTENT_RESPONSES: Dict[str, str] = {
    "greeting": "Szia! Európai tengerparti, hegyi, kulturális és kalandutakat szervezünk. Miben segíthetek?",
    "beach": "Tengerparti útból ajánlunk görög szigeteket, horvát tengerpartot, spanyol Costa Bravát vagy portugál partvidéket.",
    "mountain": "Hegyi utakhoz ajánlunk Alpok, Dolomitok, Tátra vagy Pyrenee-k környéki programokat.",
    "cultural": "Kulturális utaknál népszerű célok: Prága, Bécs, Róma, Budapest, Krakkó és Párizs.",
    "adventure": "Kalandos utaknál van rafting, via ferrata, túra, sí, kerékpár és egyes helyeken extrémebb program is.",
    "accommodation": "Szállásból van hotel, apartman, panzió, családbarát és all inclusive megoldás is.",
    "food": "Étkezésnél kérhető reggeli, félpanzió, teljes panzió vagy all inclusive. Diétás igény is jelezhető.",
    "insurance": "Utazási biztosítás kérhető külön, és több csomaghoz alapból is hozzáadható.",
    "accessibility": "Megközelíthetőség szerint tudunk repülős, buszos, vonatos és transzferes opciókat ajánlani.",
    "weather": "Az időjárás úticéltól és szezontól függ. Tengerpartra általában tavasz vége és nyár a legjobb, hegyekbe nyár vagy tél.",
    "destination": "Európában több célunk van: tengerpart, hegy, városnézés és aktív programok. Mondd meg, milyen stílus érdekel.",
    "price": "Az ár a célállomástól, szezontól, szállástól és ellátástól függ. Tudok segíteni szűkíteni az ajánlatot.",
    "duration": "Vannak 3-4 napos városnéző utak, egyhetes nyaralások és hosszabb, 10-14 napos csomagok is.",
    "booking": "Foglalni online vagy telefonon is lehet. Ha szeretnéd, végigvezetlek a foglalási lépéseken.",
    "transport": "Az utazás lehet repülővel, busszal vagy egyéni megközelítéssel. Transzfert is lehet kérni.",
    "family": "Családi utakhoz gyerekbarát szállást, rövidebb transzfert és nyugodtabb programot ajánlok.",
}


# =============================================================================
# 6) Szabályalapú gyors útvonal
# =============================================================================
# Ezt az útvonalat azért tartjuk meg, mert bizonyos kulcsszavaknál nagyon gyorsan
# és kiszámíthatóan lehet válaszolni.
RULES: List[Tuple[str, str]] = [
    (r"\b(szia|helló|hello|jó napot|üdv)\b", INTENT_RESPONSES["greeting"]),
    (r"\b(tengerpart|strand|mediterrán|görög|horvát|spanyol|portugál)\b", INTENT_RESPONSES["beach"]),
    (r"\b(hegy|alp|dolomit|túra|sí|havas)\b", INTENT_RESPONSES["mountain"]),
    (r"\b(kultúr|városnézés|múzeum|történelmi|prága|bécs|róma|párizs|krakkó)\b", INTENT_RESPONSES["cultural"]),
    (r"\b(extrém|adrenalin|rafting|bungee|via ferrata|kaland)\b", INTENT_RESPONSES["adventure"]),
    (r"\b(szállás|hotel|apartman|panzió|all inclusive)\b", INTENT_RESPONSES["accommodation"]),
    (r"\b(étkezés|reggeli|félpanzió|teljes panzió|diét|vegetáriánus)\b", INTENT_RESPONSES["food"]),
    (r"\b(biztosítás|utasbiztosítás|balesetbiztosítás)\b", INTENT_RESPONSES["insurance"]),
    (r"\b(megközelíthető|repülő|busz|vonat|transzfer)\b", INTENT_RESPONSES["accessibility"]),
    (r"\b(időjárás|idő|szezon|nyár|tél|ősz|tavasz)\b", INTENT_RESPONSES["weather"]),
    (r"\b(úticél|célország|ország|európa|ajánlasz)\b", INTENT_RESPONSES["destination"]),
    (r"\b(ár|árak|mennyibe|kedvezmény|akció|olcsó)\b", INTENT_RESPONSES["price"]),
    (r"\b(nap|hetes|hétvége|hány nap|mennyi ideig)\b", INTENT_RESPONSES["duration"]),
    (r"\b(foglal|előleg|befizet|online)\b", INTENT_RESPONSES["booking"]),
    (r"\b(eljut|oda jut|transzfer|jegy|út)\b", INTENT_RESPONSES["transport"]),
    (r"\b(család|gyerek|gyerekbarát|kisgyerek)\b", INTENT_RESPONSES["family"]),
]


def rule_based_chat(user_text: str) -> str:
    # Először a normalizált szövegen futtatjuk a mintákat.
    # Így a kis/nagybetű és az írásjelek kevésbé zavarják a felismerést.
    normalized = basic_normalize_hu(user_text)
    for pattern, response in RULES:
        if re.search(pattern, normalized):
            return response

    return (
        "Ezt most nem tudom biztosan besorolni. "
        "Tengerpart, hegy, városnézés, szállás, biztosítás vagy ár érdekel?"
    )


# =============================================================================
# 7) Hibrid chatbot
# =============================================================================
# A végső bot először a gyors szabályokat nézi,
# utána intent osztályozást használ,
# végül FAQ retrieval-t, ha bizonytalan.
class TravelChatbot:
    def __init__(self) -> None:
        self.intent_model = TravelIntentModel()
        self.retrieval = RetrievalEngine(FAQ)

    def fit(self, dataframe: pd.DataFrame) -> None:
        self.intent_model.fit(dataframe["text"].tolist(), dataframe["intent"].tolist())

    def predict_intent(self, user_text: str) -> IntentPrediction:
        return self.intent_model.predict_with_confidence(user_text)

    def reply(self, user_text: str, threshold: float = 0.45) -> str:
        normalized = basic_normalize_hu(user_text)

        # 1) Szabályalapú gyors válasz
        for pattern, response in RULES:
            if re.search(pattern, normalized):
                return response

        # 2) Intent osztályozás, ha kellő bizonyosságú az eredmény
        prediction = self.intent_model.predict_with_confidence(user_text)
        if prediction.confidence >= threshold:
            return INTENT_RESPONSES.get(prediction.intent, "Pontosítanád a kérdést?")

        # 3) Retrieval fallback, ha az intent modell bizonytalan
        return self.retrieval.answer(user_text)

    def explain_top_terms(self, top_k: int = 6) -> Dict[str, List[Tuple[str, float]]]:
        return self.intent_model.top_terms_by_intent(top_k=top_k)


# =============================================================================
# 8) Tanítás és gyors kiértékelés
# =============================================================================
chatbot = TravelChatbot()

# Stratified split: minden intent aránya közel azonos marad a train/test részekben.
X = df["text"].tolist()
y = df["intent"].tolist()
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y,
)

# A modellt külön tanítjuk a splitre, hogy legyen mérhető értékelésünk.
chatbot.intent_model.fit(X_train, y_train)

# Predikció a tesztkészleten.
y_pred = [chatbot.intent_model.predict(text) for text in X_test]

print("=== Osztályozási eredmény ===")
print(classification_report(y_test, y_pred, zero_division=0))
print("Konfúziós mátrix:")
print(confusion_matrix(y_test, y_pred))


# =============================================================================
# 9) Modell értelmezése: jellemző kifejezések intentenként
# =============================================================================
print("\n=== Legjellemzőbb kifejezések intentenként ===")
for intent, terms in chatbot.explain_top_terms(top_k=5).items():
    print(f"\n{intent.upper()}")
    for term, weight in terms:
        print(f"  {term:25s} {weight:.3f}")


# =============================================================================
# 10) Példák
# =============================================================================
# Ezek segítenek gyorsan ellenőrizni, hogy a chatbot logikája jól működik-e.
EXAMPLES = [
    "Szia!",
    "Tengerparti nyaralást keresek Görögországban.",
    "Van hegyi túra az Alpokban?",
    "Prága vagy Bécs érdekelne kulturális céllal.",
    "Kell utasbiztosítás is?",
    "Vegetáriánus étkezés megoldható?",
    "Mennyibe kerül egy egyhetes út?",
    "Hogyan lehet foglalni?",
    "Milyen az időjárás augusztusban a tengerparton?",
    "Repülővel vagy busszal lehet menni?",
    "Gyerekbarát családi utat keresek.",
]

print("\n=== Tesztpéldák ===")
for text in EXAMPLES:
    print(f"User: {text}")
    print(f"Bot : {chatbot.reply(text)}")
    print("-" * 60)


# =============================================================================
# 11) Interaktív chat loop
# =============================================================================
# Ha Colabban vagy egy terminálos környezetben futtatod, a következő függvény
# folyamatos beszélgetést tesz lehetővé. Kilépés: "exit".
def chat_loop(bot_fn) -> None:
    print("Írj üzenetet. Kilépés: exit\n")
    while True:
        user = input("Te: ").strip()
        if user.lower() == "exit":
            print("Kilépés.")
            break
        print("Bot:", bot_fn(user))
        print()


# Ha szeretnéd az interaktív módot használni, vedd ki a kommentet:
chat_loop(chatbot.reply)
```
